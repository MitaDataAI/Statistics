# Packages

```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from scipy.stats import t
import statsmodels.api as sm
from scipy.stats import bootstrap
from sklearn.utils import resample
```

# Load Data

```python
diabetes = load_diabetes()
diabetes
```

```python
df = pd.DataFrame(data = diabetes.data, columns= diabetes.feature_names)
df["target"] = diabetes.target
```

```python
df_bmi_target = df.copy()
```

```python
df_bmi_target = df_bmi_target[["bmi", "target"]]
```

```python
df_bmi_target
```

```python
X = df_bmi_target["bmi"] # Body Mass Index, a standardized measure to assess body fat based on height and weight.
Y = df_bmi_target["target"] # Represents a quantitative measure of disease progression one year after baseline.
```

# Correlation between bmi and target

```python
plt.figure()
sns.scatterplot(df_bmi_target, x = X, y = Y)
plt.show()
```

As we can see, when the body mass index increase, then there is a risk to have diabet. 
Let's confirm this linear relationship by the test.

```python
stat, p_value = pearsonr(X, Y)
print("stat and p_value of the test of correlation : ", stat, p_value)
```

As p_value < 5%, we can reject H0. It confirm that there is a linear relationship between "bmi" and "target".

# Make OLS graphic

```python
# Add constant
X_const = sm.add_constant(X)
X_const.head(5)
```

```python
# Ajustement
model = sm.OLS(Y, X_const).fit()
model
```

```python
# Values of graphic
xs = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
print(xs)
xs_const = sm.add_constant(xs)  # Intercept
```

```python
# Prediction + CI
predictions = model.get_prediction(xs_const)
ci = predictions.conf_int(alpha=0.05)        # IC pour la moyenne prédite
mean_pred = predictions.predicted_mean
```

```python
# Tracé
plt.scatter(X, Y, label='Data', alpha=0.5)
plt.plot(xs, mean_pred, color='red', label='Regression')
plt.fill_between(xs.ravel(), ci[:, 0], ci[:, 1], color='orange', alpha=0.3,
                 label='IC 95% (mean)')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regression CI 95% (statsmodels)")
plt.legend()
plt.show()
```

# Boostrap Resampling

Example

```python
# Stocker les échantillons bootstrap
bootstrap_samples = []

# Générer 100 échantillons bootstrap
for i in range(100):
    X_resample, Y_resample = resample(X_array, Y_array, replace=True, n_samples=len(X), random_state=i)
    bootstrap_samples.append((X_resample, Y_resample))
```

```python
bootstrap_samples
```

```python
models = []
x_range = np.linspace(X_array.min(), X_array.max(), 100).reshape(-1, 1)

for Xb, yb in bootstrap_samples:
    model = LinearRegression()
    model.fit(Xb, yb)
    y_pred = model.predict(x_range)
    label = "Bootstrap regression" * (i == 0) or None
    plt.plot(x_range, y_pred, color='red', alpha=0.1, label=label)
    models.append(model)

# Régression sur les données complètes (ligne moyenne)
final_model = LinearRegression().fit(X_array, Y_array)
plt.plot(x_range, final_model.predict(x_range), color='black', linewidth=2, label="Linear Regression for the original data")
sns.scatterplot(df, x = X, y = Y)

# Mise en forme
plt.title("Bootstrap regressor and the average")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

The born on red quantify our uncertidied with our prediction. As we can see, it is not very large.
