# Package

```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer
from sklearn.linear_model import LinearRegression
```

# Load

```python
df = load_diabetes()
```

```python
df
```

```python
type(df)
```

```python
print(df["data"])
```

```python
X = df["data"]
X[:2]
```

```python
Y = df["target"]
Y[:5]
```

```python
X_col = df["feature_names"]
X_col
```

```python
Y_col = "target"
```

# Convert to DataFrame

```python
df = pd.DataFrame(data = X, columns=X_col)
df[Y_col] = Y
```

```python
df.head()
```

# Exploration

```python
df.describe()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = ".1f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
corr_matrix["target"].sort_values(ascending=False)
```

# Density

```python
df.hist(figsize=(12, 8))
plt.suptitle('Task 5: Feature Distributions')
plt.show()
```

# Scatter plot

```python
df["target"]
```

```python
Y = df["target"]
X = df.drop("target", axis = 1)
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(df, x_vars = df.drop("target", axis = 1), y_vars = 'target', height=3)
plt.suptitle('Task 6: Scatter Plots of Features vs. Target')
plt.show()
```

# Normaliser X

```python
normalizer = Normalizer()
X_normal = normalizer.fit_transform(X)
```

```python
X_normal
```

# Régression linéaire

```python
reg = LinearRegression()
reg = reg.fit(X_normal, Y)
reg
```

# Coefficients

```python
print("intercept :", reg.intercept_, ", coefficients:", reg.coef_)
```

# Valeur prédite

```python
Y_pred = reg.predict(X_normal)
print(Y_pred[:5])
```
