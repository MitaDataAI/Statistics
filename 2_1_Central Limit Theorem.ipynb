```python
import numpy as np
import matplotlib.pyplot as plt
```

# Demonstrate the Central Limit Theorem

We will generate 10000 samples of differentes dimensions (m_size). After that, we will plot the mean.  

```python
# Parameters
m_size = [2, 20, 50, 1000]  # Different sample sizes to explore
n = 10000  # Number of simulations for each sample size
# We will generate then (m_values * n) samples
```

```python
# Create a figure to display histograms
plt.figure(figsize=(12,8))

# Loop through different sample sizes
for m in m_size :
    samples_mean = []

    # Simulate the Central Limit Theorem
    for _ in range(n) :
        # Generate random samples from a uniform distribution
        random_samples = np.random.uniform(0,1,m)
        
        # Calculate the sample mean
        sample_mean = np.mean(random_samples)
        samples_mean.append(sample_mean)

    # Create a subplot for each sample size
    plt.subplot(2,2,m_size.index(m) + 1)

    plt.hist(samples_mean, bins = 50, alpha = 0.6)

    plt.title(f'Sample Size (m) = {m}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')

# Adjust subplot layout
plt.tight_layout()

# Display the plots
plt.show()
```
